```python
import os
from google.colab import drive

# 1. Hubungkan ke Google Drive
drive.mount('/content/drive')

# 2. Tentukan nama file backup dan folder kerja
BACKUP_PATH = "/content/drive/MyDrive/RWKV_Project/rwkv_v7_env.tar.gz"
# SUDAH DISESUAIKAN DENGAN NAMA REPO ANDA
REPO_URL = "https://github.com/miftahul001/RWKV-V7.git"

# 3. Clone Repositori dari GitHub Anda (jika belum ada)
# Folder hasil clone akan bernama RWKV-V7
if not os.path.exists('/content/RWKV-V7'):
    print("📥 Cloning repositori dari GitHub...")
    !git clone {REPO_URL}

# 4. LOGIKA INTEGRASI: Cek Backup di Drive
if os.path.exists(BACKUP_PATH):
    print("🚀 Backup ditemukan di Drive! Mengekstrak environment Python 3.12...")
    !tar -xzvf {BACKUP_PATH} -C /usr/local/envs
    print("✅ Environment siap digunakan.")
else:
    print("⚠️ Backup tidak ditemukan. Memulai instalasi bersih (ini akan memakan waktu)...")

    # Buat environment baru
    !conda create -n rwkv_env python=3.12 -y

    # Instal requirements dari folder hasil clone tadi (PATH SUDAH DISESUAIKAN)
    print("📦 Menginstal dependensi dari requirements.txt...")
    !conda run -n rwkv_env pip install -r /content/RWKV-V7/requirements.txt

    # Simpan ke Drive agar besok tidak instal lagi
    print("💾 Menyimpan hasil instalasi ke Google Drive untuk penggunaan masa depan...")
    !tar -czvf {BACKUP_PATH} -C /usr/local/envs rwkv_env
    print("✅ Selesai! Backup tersimpan di Drive.")

# 5. TEST: Cek apakah environment sudah aktif
print("\nDaftar Environment:")
!conda env list
```
